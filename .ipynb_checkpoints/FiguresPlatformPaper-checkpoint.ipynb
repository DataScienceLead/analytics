```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../Data/simulatedData.xlsx')
df_prob = pd.read_excel('../Data/simulatedData.xlsx' , sheet_name='Sheet2')
                        
```

```python
df_prob.head()
```

```python
df.sample()
```

```python
folderfig = '../'
```

```python
# hue='Agent'

df_act = df[df['Agent']!= 'Platform'].copy()

fig, ax = plt.subplots()
sns.kdeplot(data = df_act, x = 'CreditScore', hue='Agent',     common_norm=False, common_grid=True,)

ax.set(xlabel='Score, internal credit model')
filename = folderfig+'kdeActual.tiff'
plt.savefig(filename)

plt.show()
```

```python
df_pla = df[df['Agent']!= 'Actual credit risk'].copy()

fig, ax = plt.subplots()
sns.kdeplot(data = df_pla, x = 'CreditScore', hue='Agent',     common_norm=False, common_grid=True,)

ax.set(xlabel='Score, internal credit model')
filename = folderfig + 'kdePlatform.tiff'
plt.savefig(filename)

plt.show()
```

```python
# df_prob.plot()
fig, ax = plt.subplots()
plt.plot(df_prob)

ax.set(xlabel='Standard deviations')
ax.set(ylabel='Probability')
ax.legend( ['0', '1', '2', '4', '7', '14'])
# plt.ylabel('Probability')

filename = folderfig + 'prob.tiff'
plt.savefig(filename)
plt.show()
```
